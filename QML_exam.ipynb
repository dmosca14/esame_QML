# QSVM per ogni dataset

L'importazione dei dataset avviene in due fasi. Come prima cosa, generiamo o carichiamo (in caso siano dati reali) i dataset, nella forma dataset = (X_dataset, y_dataset), dove X_dataset sono i campioni e y_dataset i relativi target, che possono essere 0 oppure 1. Successivamente, il dataset viene splittato fra train set e test set; l'output della funzione che fa questo lavoro viene chiamato "set" e, in generale, in questo codice qualsiasi set è una tupla della forma set (train_set, test_set) dove, a loro volta, train_set = (X_train_set, y_train_set) e test_set = (X_test_set, y_test_set). DATASET è X E Y, SET è QUANDO è SPLITTATO.
