# Import

```python
from sklearn.datasets import make_classification, make_circles, make_moons
from sklearn.datasets import load_breast_cancer

import numpy as np

from modules.quantum_devices import lightning_qubit_device, allocate_GPU_device
from modules.kernels import kernel_amplitude, kernel_angle, kernel_basis, kernel_IQP, kernel_projected, kernel_rbf
from modules.main import qsvm
```

# General configuration

```python
config = {"number_samples": 200,
          "number_features": 2,
          "fraction_train": 0.80,
          "max_qubits": 4,
          "quantum_device": lightning_qubit_device}

kernel_modules = [kernel_rbf, # kernel_rbf MUST BE the first kernel of the list.
                  kernel_amplitude,
                  kernel_angle,
                  kernel_basis,
                  kernel_IQP,
                  kernel_projected]
```

# Linear dataset

```python
dataset_linear_data = make_classification(
        n_samples = config["number_samples"],
        n_features = config["number_features"], 
        n_redundant = 0, 
        n_informative = 2, 
        random_state = 42, 
        n_clusters_per_class = 1
        )

dataset_linear = {"dataset_name": "linear", "dataset_original_data": dataset_linear_data}

qsvm(dataset_linear, config, kernel_modules)
```

# Circles dataset

```python
dataset_circles_data = make_circles(
        n_samples = config["number_samples"], 
        noise = 0.08, 
        factor = 0.3, 
        random_state = 42
        )

dataset_circles = {"dataset_name": "circles", "dataset_original_data": dataset_circles_data}

qsvm(dataset_circles, config, kernel_modules)
```

# Moons dataset

```python
dataset_moons_data = make_moons(
        n_samples = config["number_samples"],
        noise = 0.1,
        random_state = 42
        )

dataset_moons = {"dataset_name": "moons", "dataset_original_data": dataset_moons_data}

qsvm(dataset_moons, config, kernel_modules)
```

# "Quantum Ad Hoc" dataset

```python
np.random.seed(42)
X_dataset_quantum_ad_hoc = np.random.uniform(-np.pi, np.pi, (config["number_samples"], 2))
y_dataset_quantum_ad_hoc = np.array([1 if np.sin(x[0]) * np.cos(x[1]) > 0 else 0 for x in X_dataset_quantum_ad_hoc])

dataset_quantum_ad_hoc_data = (X_dataset_quantum_ad_hoc, y_dataset_quantum_ad_hoc)

dataset_quantum_ad_hoc = {"dataset_name": "quantum_ad_hoc", "dataset_original_data": dataset_quantum_ad_hoc_data}

qsvm(dataset_quantum_ad_hoc, config, kernel_modules)
```

# High dimensional XOR dataset

```python
rng = np.random.RandomState(42)
X_high_dim_XOR = rng.uniform(-1, 1, (config["number_samples"], config["number_features"]))
y_high_dim_XOR = (np.prod(np.sign(X_high_dim_XOR), axis=1) > 0).astype(int)

dataset_high_dim_XOR_data = (X_high_dim_XOR, y_high_dim_XOR)

dataset_high_dim_XOR = {"dataset_name": "high_dim_XOR", "dataset_original_data": dataset_high_dim_XOR_data}

qsvm(dataset_high_dim_XOR, config, kernel_modules)
```
